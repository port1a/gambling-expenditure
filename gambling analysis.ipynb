```python
# Some basic data wrangling from mpesa statements to obtain analysis of total transactions gambling with various Kenyan betting sites over the years
```

```python
import pandas as pd
```

```python
mpesa=pd.read_excel("C:\\Users\\STACY\\Documents\\mpesa 9.xlsx")
```

```python
mpesa.head(15)
```

```python
mpesa.drop(columns=['Unnamed: 6', "Transaction Status"], inplace=True)
```

```python
mpesa.dropna(axis=0, inplace=True)
```

```python
mpesa
```

```python
betway=mpesa["Details"].str.contains('Betway') # filter for columns with transactions sent to Betway
```

```python
frmbetway=mpesa["Details"].str.contains('BETWAY') # filter for columns with transactions paid out from Betway
```

```python
betpawa=mpesa['Details'].str.contains('GAMING') # filter for columns with transactions sent to and from Betpawa
```

```python
odibets=mpesa['Details'].str.contains('ODIBETS') # filter for columns with transactions sent to and from Odibets
```

```python
sportpesa=mpesa['Details'].str.contains('SportPesa') # filter for columns with transactions sent to and from Sportpesa
```

```python
betin=mpesa["Details"].str.contains('BETIN') # filter for columns with transactions sent to and from Betin
```

```python
betika=mpesa["Details"].str.contains("BETIKA") # filter for columns with transactions sent to and from Betika
```

```python
# create masks to filter through data
bets1=mpesa[betway]
bets2=mpesa[betika]
bets3=mpesa[betpawa]
bets4=mpesa[betin]
bets5=mpesa[odibets]
bets6=mpesa[sportpesa]
bets7=mpesa[frmbetway]
```

```python
# Create dataframe after masking data
betts=pd.concat([bets1,bets2,bets3,bets4,bets5,bets6, bets7],ignore_index=True)
```

```python
betts
```

```python
# Export ddataframe as Excel file 
betts.to_excel('C:\\Users\\STACY\\Documents\\gambler.xlsx')
```
